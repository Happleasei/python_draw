### Python画图总结 

```python
import pandas as pd
data = pd.read_csv("./data/data_analyze.csv")
```

```python
data.head()
```

#### 折线图 

```python
# 折线图
import matplotlib.pyplot as plt
#用来正常显示中文标签
plt.rcParams['font.sans-serif']=['SimHei']
#用来正常显示负号
plt.rcParams['axes.unicode_minus']=False
# 画布的尺寸
plt.figure(figsize=(25, 4))
# 
plt.xticks(size=14, rotation=60, color="white")
plt.yticks(size=14, color="white")
# 时间列包含秒，导致横坐标太密，适当优化
data["时间"] = data["时间"].map(lambda x: x.split(":")[0])
plt.plot(data["时间"], data["一次风转速"], color="red", label="一次风转速")
plt.legend()
plt.savefig("./imgs/3.png", bbox_inches="tight")
plt.show()
```

```python
x = [i for i in range(len(data["时间"]))]
l_c = len(data.columns)
# 画布的尺寸
plt.figure(figsize=(25, 3*(l_c-1)))
for i in range(1, l_c):
    plt.subplot(l_c, 1, i+1)
    plt.yticks(size=14, color="white")
    plt.xticks(size=14, color="white")
    plt.plot(x, data[data.columns[i]], label=data.columns[i])
    plt.legend()
plt.savefig("./imgs/4.png", bbox_inches="tight")
plt.show()
```

```python
x = [i for i in range(len(data["时间"]))]
l_c = len(data.columns)
# 画布的尺寸
plt.figure(figsize=(6*(l_c-1), 5))
for i in range(1, l_c):
    plt.subplot(1, l_c, i+1)
    plt.xlabel(data.columns[i], size=14, color="white")
    plt.ylabel("取值统计", size=14, color="white")
    plt.yticks(size=14, color="white")
    plt.xticks(size=14, color="white")
    plt.plot(x, data[data.columns[i]], label=data.columns[i])
    plt.legend()
plt.savefig("./imgs/5.png", bbox_inches="tight")
plt.show()
```

#### 饼图 

```python
# 画布的尺寸
data = {
    '中国': (130, '#7199cf'),
    '美国': (115, '#4fc4aa'),
    '日本': (60, '#ffff10'),
}
cities = data.keys()
values = [x[0] for x in data.values()]
colors = [x[1] for x in data.values()]
labels = ['{}:{}'.format(city, value) for city, value in zip(cities, values)]
explode = [0.05, 0, 0] # 设置饼图的凸出显示
plt.figure(figsize=(5, 5))
plt.xlabel("比例", size=14, color="white")
plt.ylabel("取值统计", size=14, color="white")
plt.yticks(size=14, color="white")
plt.xticks(size=14, color="white")
plt.pie(values, labels=labels, colors=colors, explode=explode, shadow=True) # 画饼状图， 并且指定标签和对应的颜色  指定阴影效果
plt.legend()
plt.savefig("./imgs/13.png", bbox_inches="tight")
plt.show()
```

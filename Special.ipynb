```python
import pandas as pd
data = pd.read_csv("./data/data_analyze.csv")
```

```python
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#用来正常显示中文标签
plt.rcParams['font.sans-serif']=['SimHei']
#用来正常显示负号
plt.rcParams['axes.unicode_minus']=False
plt.figure(figsize=(12, 6))
plt.xticks(size=16, color="white")
plt.yticks(size=16, color="white")
sns.heatmap(data.corr(method="spearman"), annot=True, annot_kws={'size':19,'weight':'bold'})
plt.savefig("./imgs/14.png", bbox_inches="tight")
```
